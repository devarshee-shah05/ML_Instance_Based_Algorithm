# SVM = Support Vector Machine

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head(15)
```

```python
df.shape
```

# Target and non-target columns

```python
X = df.iloc[:,[2,3]].values
Y = df.iloc[:,-1].values
```

# Train and test data set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=50)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
var_StandardScaler = StandardScaler()
X_train = var_StandardScaler.fit_transform(X_train)
X_test = var_StandardScaler.transform(X_test)
```

# Applying svm model

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear',random_state = 0)
classifier.fit(X_train, Y_train)

# Predicting the test set results
Y_Pred = classifier.predict(X_test)
```

```python
Y_Pred
```

```python
from sklearn import metrics
print('Accuracy Score: with linear kernel')
print(metrics.accuracy_score(Y_test,Y_Pred))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf')
classifier.fit(X_train,Y_train)

# Predicting the test set results
Y_pred = classifier.predict(X_test)
print('Accuracy Score: with default rbf kernel')
print(metrics.accuracy_score(Y_test,Y_Pred))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf',gamma=15,C=7, random_state=0)
classifier.fit(X_train,Y_train)

# Predicting the test set results
Y_Pred = classifier.predict(X_test)

print('Accuracy Score On Test Data: with default rbf kernel')
print(metrics.accuracy_score(Y_test,Y_Pred))
```
